```python
%matplotlib inline
import zipfile
import numpy as np
import pandas as pd
import seaborn as sns

z = zipfile.ZipFile(r"C:\Users\user\Desktop\popythonilo\PyLess-main\athlete_events.zip")
df = pd.read_csv(z.open("athlete_events.csv"))
df = df.dropna(subset=['Medal', "Age", "Height", "Weight"])
df.head()
```

**1. Сколько мужчин и женщин получили золотые, серебрянные и бронзовые медали?**

```python
# Your code here
medal_counts = df.groupby(['Sex', 'Medal']).size().reset_index(name='Count')
print(medal_counts)
```

**2. Какая страна получила наибольшее количество золотых медалей за всю историю олимпийских игр?**

```python
# Your code here
top_gold = df[df['Medal'] == 'Gold'].groupby('NOC').size().idxmax()
print(top_gold)
```

**3. Выведите распределение пола участника олимпиады от вида спорта (crosstab)**

```python
#Распределение пола от вида спорта
gender_sport_crosstab = pd.crosstab(df['Sport'], df['Sex'])
print(gender_sport_crosstab)
```

**4. Выведите средний возраст и его стандартное отклонения для женщин, учавствовавших в хоккее на льду**

```python
#Средний возраст и стандартное отклонение для женщин в хоккее на льду
ice_hockey_women = df[(df['Sport'] == 'Ice Hockey') & (df['Sex'] == 'F')]
mean_age = ice_hockey_women['Age'].mean()
std_dev_age = ice_hockey_women['Age'].std()
print(f"Средний возраст: {mean_age}, Стандартное отклонение: {std_dev_age}")
```

**5. У какой страны больше всего было больше всего женщин, получивших бронзовую медаль?**

```python
#Страна с наибольшим количеством бронзовых медалей среди женщин
top_bronze_women = df[(df['Sex'] == 'F') & (df['Medal'] == 'Bronze')].groupby('NOC').size().idxmax()
print(top_bronze_women)
```

**6. Постройте гистограмму распределения количества медалей (бронза, серебро, золото) для первых трех стран, получивших наибольшее количество медалей**

```python
#Гистограмма для топ-3 стран по количеству медалей
from matplotlib import pyplot as plt

top_countries = df['NOC'].value_counts().nlargest(3).index
data_top_countries = df[df['NOC'].isin(top_countries)]
sns.countplot(x='NOC', hue='Medal', data=data_top_countries)
plt.show()
```

**7. Нарисуйте распределение веса мужчин, получивших серебрянную медаль(density или distplot)**

```python
#Распределение веса мужчин с серебряной медалью
from matplotlib import pyplot as plt

mens_silver_weight = df[(df['Sex'] == 'M') & (df['Medal'] == 'Silver')]['Weight']
sns.displot(mens_silver_weight, kde=True)
plt.show()
```

**8. Постройте boxplot для возраста участника в зависимости от медали**

```python
#Боксплот для возраста по видам медалей
from matplotlib import pyplot as plt

sns.boxplot(x='Medal', y='Age', data=df)
plt.show()
```

**9. Постройте pairplot для веса, возраста и роста участников от USA.**

```python
#Pairplot для участников из США по весу, возрасту и росту
from matplotlib import pyplot as plt


usa_participants = df[df['NOC'] == 'USA'][['Weight', 'Age', 'Height']]
sns.pairplot(usa_participants)
plt.show()
```
