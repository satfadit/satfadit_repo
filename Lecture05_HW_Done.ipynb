**1. Нормализовать данные**

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df_scaled = pd.DataFrame(scaler.fit_transform(df.iloc[:, :-4]), columns=df.columns[:-4])
df_scaled = pd.concat([df_scaled, df.iloc[:, -4:]], axis=1)
```

**2. Построить бинарный классификатор для каждого семейства(Family) на основе LogisticRegression c $l_1$ и $l_2$ регуляризацими. Это соответствует целевым исходам (Конкретное семейство, Всё остальное). Подобрать гиперпараметр $C$ с помощью кросс-валидации, используя метрику ROC AUC.**

```python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_auc_score

# Подготовка данных
X = df_scaled.iloc[:, :-4]
y_families = df_scaled['Family']

families = df['Family'].unique()
results = {}

for family in families:
    y_bin = (y_families == family).astype(int)
    X_train, X_test, y_train, y_test = train_test_split(X, y_bin, test_size=0.2, random_state=42)

    param_grid = {'C': [0.01, 0.1, 1, 10], 'penalty': ['l1', 'l2']}
    grid = GridSearchCV(LogisticRegression(solver='liblinear'), param_grid, cv=5, scoring='roc_auc')
    grid.fit(X_train, y_train)

    model = grid.best_estimator_
    y_pred_proba = model.predict_proba(X_test)[:, 1]
    roc_auc = roc_auc_score(y_test, y_pred_proba)

    results[family] = {
        'Best Parameters': grid.best_params_,
        'ROC AUC': roc_auc
    }

# Выведем результаты
for family, data in results.items():
    print(f"Family: {family}, Best Params: {data['Best Parameters']}, ROC AUC: {data['ROC AUC']}")
```

**3. Добавить полиномиальные признаки до второго порядка. Построить бинарный классификатор для каждого семейства на основе LogisticRegression c $l_1$ и $l_2$ регуляризацими. Подобрать гиперпараметр $C$ с помощью кросс-валидации, используя метрику ROC AUC.**

```python
from sklearn.preprocessing import PolynomialFeatures

poly = PolynomialFeatures(2)
X_poly = poly.fit_transform(X)

results_poly = {}
for family in families:
    y_bin = (y_families == family).astype(int)
    X_train, X_test, y_train, y_test = train_test_split(X_poly, y_bin, test_size=0.2, random_state=42)

    param_grid = {'C': [0.01, 0.1, 1, 10], 'penalty': ['l1', 'l2']}
    grid = GridSearchCV(LogisticRegression(solver='liblinear'), param_grid, cv=5, scoring='roc_auc')
    grid.fit(X_train, y_train)

    model = grid.best_estimator_
    y_pred_proba = model.predict_proba(X_test)[:, 1]
    roc_auc = roc_auc_score(y_test, y_pred_proba)

    results_poly[family] = {
        'Best Parameters': grid.best_params_,
        'ROC AUC': roc_auc
    }

# Выведем результаты для полиномиальных признаков
for family, data in results_poly.items():
    print(f"Family (Poly): {family}, Best Params: {data['Best Parameters']}, ROC AUC: {data['ROC AUC']}")
```
